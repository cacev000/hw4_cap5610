```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

pos = np.array([[1,1], [2,2], [2,0]])
neg = np.array([[0,0], [1,0], [0,1]])

plt.scatter(pos[:,0], pos[:,1], color="green", label="Pos")
plt.scatter(neg[:,0], neg[:,1], color="red", label="Neg")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("x1 vs x2")
plt.show()
```
